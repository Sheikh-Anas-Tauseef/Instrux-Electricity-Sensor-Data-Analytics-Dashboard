```python
import pandas as pd
```

```python
df = pd.read_csv(
    "instrux_135_jan_feb.csv",
    engine="python",
    on_bad_lines="skip"
)
```

```python
print(df.shape)   # rows, columns
df.head()
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df.head()
```

```python
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print(constant_cols)
```

```python
df_clean = df.drop(columns=constant_cols)
```

```python
df_clean.dtypes
```

```python
df_clean["datetime"] = pd.to_datetime(df_clean["datetime"])
df_clean.set_index("datetime", inplace=True)
```

```python
df_clean.drop(columns=["datetime_server"], inplace=True)
```

```python
df_clean.info()
df_clean.head()
```

```python
start_date = df_clean.index.min()
end_date = df_clean.index.max()

print("Start date:", start_date)
print("End date:", end_date)
```

```python
df_clean.to_csv('clean.csv')
```
